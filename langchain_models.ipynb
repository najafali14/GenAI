```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

google_api_key= 'AIzaSyAxz3kNZLBz2PH124b-pfqVuulj960QvKo'
embedding = GoogleGenerativeAIEmbeddings(model='models/embedding-001',google_api_key=google_api_key, dimensions=300)

documents = [
    "Babar Azam is a Pakistani cricketer known for his elegant batting and consistency.",
    "Shaheen Afridi is a Pakistani fast bowler famous for his pace and swing.",
    "Mohammad Rizwan is a wicketkeeper-batsman known for his resilience and match-winning knocks.",
    "Shahid Afridi, also known as 'Boom Boom Afridi', was famous for his explosive batting and leg-spin.",
    "Wasim Akram, regarded as one of the greatest fast bowlers, was known for his lethal swing and reverse swing."
]


query = 'tell me about wasim akram'

doc_embeddings = embedding.embed_documents(documents)
query_embedding = embedding.embed_query(query)

scores = cosine_similarity([query_embedding], doc_embeddings)[0]

index, score = sorted(list(enumerate(scores)),key=lambda x:x[1])[-1]

print(query)
print(documents[index])
print("similarity score is:", score)

```
